```python
# Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import the Zomato dataset directly from the ZIP file using pandas
# 'latin-1' encoding is used to avoid errors from special characters
# 'compression="zip"' tells pandas the file is inside a zip archive
df = pd.read_csv("C:/Users/Vishal Kumar Sharma/Downloads/zomato.csv.zip", encoding='latin-1', compression='zip')
# Display the first 5 rows of the dataset to get an overview
df.head()
```

```python
# Provide a conside summary of the DataFrame, including the number of non-null entries and data types for each column.
df.info()
```

```python
# Provides the data type of each column in the DataFrame
df.dtypes
```

```python
# Data Cleaning 
#Removes unnecessary or irrelevant columns from the dataset.
# Drop irrelevant columns
df.drop(['url', 'address', 'phone', 'dish_liked', 'reviews_list', 'menu_item'], axis=1, inplace=True)

# Rename columns for simplicity
df.rename(columns={'approx_cost(for two people)': 'cost', 'listed_in(type)': 'service_type', 'listed_in(city)': 'city'}, inplace=True)

# Drop duplicates
df.drop_duplicates(inplace=True)

# Check missing values
df.isnull().sum()
```

```python
# Drop rows with missing values in key columns
df = df.dropna(subset=['rate', 'location', 'rest_type', 'cuisines'])

# Fill missing values in remaining columns if needed
df['cost'] = df['cost'].fillna(df['cost'].mode()[0])
```

```python
# Clean 'rate'column 
df['rate'] = df['rate'].apply(lambda x:str(x).split('/')[0].strip() if '/' in str(x) else np.nan)
df['rate'] = df['rate'].replace('NEW', np.nan).replace('-',np.nan)
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')
```

```python
# Convert cost to numeric
df['cost'] = df['cost'].astype(str).str.replace(',','')
df['cost'] = pd.to_numeric(df['cost'],errors='coerce')
```

```python
# Encode categorical features
df['online_order'] = df['online_order'].map({'Yes':1, 'No':0})
df['book_table'] = df['book_table'].map({'Yes': 1, 'No': 0})
```

```python
# Generates summary statistics for numeric columns, including count, mean and quartiles.
df.describe
```

```python
# Pattern, Trends & Anomalies
# Top cuisines
print(df['cuisines'].value_counts().head(10))
```

```python
# Average rating by city
print(df.groupby('city')['rate'].mean().sort_values(ascending = False).head())
```

```python
# Outliers in cost
plt.figure(figsize=(8,4))
sns.boxplot(df['cost'])
plt.title('cost for Two -Outlier Detection')
plt.show()
```

```python
# Visual Representations of DataSet
# Online Order Availability of  Visual Representations
sns.countplot(x='online_order', data = df)
plt.title('Online Order Availability')
plt.xlabel('Online Order')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
# Table Booking Availability  of Visual Representations
sns.countplot(x='book_table', data = df)
plt.title('Table Booking Availability')
plt.xlabel('Book Table')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
# Cost for Two Distribution of Visual  Representations
sns.histplot(df['cost'].dropna(), bins=30, kde=True)
plt.title('Cost for Two Distribution')
plt.xlabel('Cost')
plt.ylabel('Count')
plt.show()
```

```python
# Top 10 Restaurant Types
df['rest_type'].value_counts().head(10).plot(kind='barh', color='skyblue')
plt.title('Top 10 Restaurant Types')
plt.xlabel('Number of Restaurants')
plt.ylabel('Restaurant Type')
plt.show()
```

```python
# Top 10 Cuisines
df['cuisines'].value_counts().head(10).plot(kind='bar', color='orange')
plt.title('Top 10 Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Most Restaurants by Location 
df['location'].value_counts().head(10).plot(kind='bar', color='green')
plt.title('Top 10 Restaurant Locations')
plt.xlabel('Location')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()
```

```python
# Boxplot – Cost by Restaurant Type
top_rest_types = df['rest_type'].value_counts().nlargest(5).index
plt.figure(figsize=(10, 5))
sns.boxplot(x='rest_type', y='cost', data=df[df['rest_type'].isin(top_rest_types)])
plt.title('Cost by Top Restaurant Types')
plt.xlabel('Restaurant Type')
plt.ylabel('Cost for Two')
plt.xticks(rotation=45)
plt.show()
```

```python
#  Restaurant Count by Cost Range (Histogram with Bins)
plt.figure(figsize=(8, 5))
sns.histplot(df['cost'], bins=[0, 200, 400, 600, 800, 1000, 2000], kde=False)
plt.title('Restaurant Count by Cost Range')
plt.xlabel('Cost for Two')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
# Restaurant Count by Cuisine 
plt.figure(figsize=(10, 5))
df['cuisines'].value_counts().nlargest(10).plot(kind='bar', color='teal')
plt.title('Top 10 Cuisines')
plt.xlabel('Cuisines')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()
```
